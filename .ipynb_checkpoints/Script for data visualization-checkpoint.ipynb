Code to read the data from pickled data with pickle library

```python
def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import TensorDataset, DataLoader
import torch.optim as optim
```

# Creating the Dataloader from the dataset #

```python
bs = 8

x_train= x_train.reshape(50000, 3, 32, 32)
x_test= x_test.reshape(10000, 3, 32, 32)

x_train, y_train, x_test, y_test = map(torch.tensor, (x_train, y_train, x_test, y_test))

train_ds = TensorDataset(x_train/255.0, y_train)
test_ds = TensorDataset(x_test/255.0, y_test)

train_dl = DataLoader(train_ds, batch_size=bs, shuffle=True)
test_dl = DataLoader(test_ds, batch_size=bs*2)
```

## Model ##

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=6, kernel_size=5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)
        self.fc1 = nn.Linear(16 * 5 * 5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 16 * 5 * 5)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

```

## Setting up the loss function and optimizer##

```python
model = Net()
loss_func = nn.CrossEntropyLoss()
opt = optim.SGD(model.parameters(), lr=1e-3, momentum=0.85)

epochs = 10
```

## Training Loop ##

```python
for epoch in range(epochs):
    train_loss = 0.0
    model.train()
    for xb, yb in train_dl:
        preds = model(xb)
        loss = loss_func(preds, yb)
        train_loss +=loss
        
        loss.backward()
        opt.step()
        opt.zero_grad()
    
    
    model.eval()
    with torch.no_grad():
        valid_loss = sum(loss_func(model(xb), yb) for xb, yb in test_dl)
    print('Epoch->', epoch, 
          ' Train Loss: ', (train_loss/len(train_dl)).item(),  
          ' Valid Loss:', (valid_loss/len(test_dl)).item())
```

```python
# Calculating the accuracy of the model
def accuracy(model_in, test_dl):
    correct = 0.0
    batch_no =1
    for xb, yb in test_dl:
        
        preds = torch.argmax(model_in(xb), dim=1)
        correct += (preds==yb).float().sum()
#         print(batch_no, len(xb), (preds==yb).float().sum())
        batch_no+=1
    return correct/10000
    
```

```python
accuracy(model, test_dl)
```

# Showing Predictions #

```python
def show_predictions(model, test_dl, batches=1):
    dataiter = iter(test_dl)
    images, labels = dataiter.next()
    preds = torch.argmax(model(images), axis =1).tolist()
    images = np.array(images).transpose(0,2,3,1)
    rows = 4
    cols = 4
    index = 0
    plt.figure(figsize=(cols+3, rows+3))
    for i in range(rows):
        for j in range(cols):
            plt.subplot(rows, cols, index+1)
            plt.imshow(images[index])
            plt.title(str(classes[labels[index].item()]) +'/' + classes[preds[index]] )
            plt.axis(False)
            index +=1
    
show_predictions(model, test_dl, batches=1)
```
